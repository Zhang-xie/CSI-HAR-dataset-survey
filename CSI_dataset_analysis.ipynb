```python
def sample_class_plot(labels:list):
    # plot a bar chart of the distribution of samples' volume on class
    distribution = Counter(labels)
    fig= plt.figure(figsize=(14,8))
    plt.bar(distribution.keys(),distribution.values())
    plt.xlabel('class label')
    plt.ylabel('No. samples')
    plt.title('Distribution of samples on class')
    plt.show()
    
```

```python
from tqdm import tqdm
def sample_len_plot(dataset):
    sample_frame = []
    for sample in tqdm(dataset):
        sample_frame.append(sample['raw'].shape[0])
    distribution = Counter(sample_frame)
    fig= plt.figure(figsize=(14,8))
    plt.bar(distribution.keys(),distribution.values())
    plt.xlabel('No. frames')
    plt.ylabel('No. samples')
    plt.title('Distribution of samples on frames')
    plt.show()
```

```python
def visulization(dataset):
    print("the number of samples:", len(dataset))
    sample_class_plot(dataset.act_label)
    for i,sample in enumerate(dataset):
        print('the types of labels:',sample.keys())
        print('the type {} and shape {} of CSI sample.'.format(type(sample['raw']),sample['raw'].shape))
        break
    sample_len_plot(dataset)
    print("Activity labels:", set(dataset.act_label), type(dataset.act_label))
```
